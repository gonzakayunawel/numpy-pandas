```python
import pandas as pd
```

```python
df_books = pd.read_csv('./data/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv', sep=',', header=0)
df_books.head(2)
```

### Drop columns

```python
# Usamos el comando eliminamos la columna: .drop('[Nombre Columna]', axis=1)
# esto en realidad es mostrar el dataframe sin la columna indicada
# axis=1: columnas
# axis=0: filas

df_books.drop('Genre', axis=1).head(2)
```

```python
# Si queremos eliminarla del dataframe, tendremos que aÃ±adir el parÃ¡metro
# inplace=True

df_books.drop('Genre', axis=1, inplace=True)
```

```python
df_books.head(2)
```

```python
# Otra manera es reasignar el valor del dataframe con la columna eliminada
df_books = df_books.drop('Year', axis=1)
```

```python
# Otra manera es eliminar la columna usando la funciÃ³n built-in de Python

del df_books['Price']
```

```python
df_books
```

## Drop rows

```python
df_books.head(2)
```

```python
# En ves de indicar el nombre de la columna, tendremos que indicar el Ã­ndice de fila
# Y en el caso del axis se coloca 0 para indicar la fila (0 fila, 1 columna)

df_books.drop(0, axis=0).head(2)
```

```python
# Podemos eliminar una lista
# Como no hems usado inplace los cambios no se han guardado

df_books.drop([0,1,2], axis=0).head(2)
```

```python
# TambiÃ©n nos permite usar range() para elminar un rango de filas
# incluso podemos usar el step para eliminar cada cierto nÃºmero de filas

df_books.drop(range(0,10,2), axis=0).head(100)
```

## Add columns

```python
df_books.head(2)
```

```python
# Para agregar una nueva columna, usamos la notaciÃ³n de llamar al dataframe y nombrar un nueva columna
# luego le damos un valor, para ello nos apoyamos con Numpy (np.nan: crea un valor NaN o no numÃ©rico (Not a Number))

import numpy as np

df_books['Nueva Columna'] = np.nan
```

```python
df_books
```

```python
# Ahora podemos crear un arange con Numpy llamado data

data = np.arange(0,df_books.shape[0])
```

```python
# Y su valor agregarlo al dataframe como una nueva columna

df_books['Rango'] = data
```

```python
# La regla es que la nueva columna debe tener la misma cantidad de filas que las ya existentes
# Para saber la cantidad nos podemos apoyar con [nombre dataframe].shape[0]
df_books
```

## Add rows

```python
# En Pandas hay una funciÃ³n similar al append de las listas que se llama .append()
# Se utilizar para agregar filas o registros

# Ejemplo:
# Agregar el mismo df, esto nos creo un duplicado agregado al df
df_books.append(df_books)
```
