```python
df_books.head(20)
```

## Apply

```python
# Definimos una función para ser aplicada

def two_times(value):
    return value * 2
```

```python
# Usamos la función apply usando como argumento el nombre de la función (si realizar un call, o sea, sin paréntesis)
# Le aplica la función a cada uno de los valores que compone a las filas que posee la columna solicitada

df_books['User Rating'].apply(two_times)
```

```python
# Podemos usar apply para crear nuevas columnas

df_books['Rating 2'] = df_books['User Rating'].apply(two_times)
```

```python
df_books
```

```python
# También podemos usar lambda functions

df_books['Rating 3'] = df_books['User Rating'].apply(lambda x: x*3)
```

```python
df_books.head(20)
```

```python
# Podemos usar distintas columnas para hacer nuestras operaciones
# Acá usamos una condición para aplicar la lambda function
# Lo que no tenemos que olvidar es especificar el eje al cual estamos referenciando
# Si no hacemos esto nos dará un error

df_books['Rating 4'] = df_books.apply(lambda x: x['User Rating'] * 2 if x['Genre'] == 'Fiction' else x['User Rating'], axis=1)
```

```python
df_books.head(20)
```

### Ventajas de Apply

* Es mucho más rápido que hacerlo mediante for
* Es posible llamar a alguna api o recurso usando una funcion y de esa manera rellenar los datos
