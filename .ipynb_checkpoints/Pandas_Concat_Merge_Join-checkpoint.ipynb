```python
import pandas as pd
import numpy as np
```

# Dataframes de prueba

```python


df1 = pd.DataFrame({'A':['A0', 'A1', 'A2', 'A3'],
'B':['B0', 'B1', 'B2', 'B3'],
'C':['C0', 'C1', 'C2', 'C3'],
'D':['D0', 'D1', 'D2', 'D3']})
df1
```

```python
df2 = pd.DataFrame({'A':['A4', 'A5', 'A6', 'A7'],
'B':['B4', 'B5', 'B6', 'B7'],
'C':['C4', 'C5', 'C6', 'C7'],
'D':['D4', 'D5', 'D6', 'D7']})
df2
```

## Concat

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], ignore_index=True)
```

```python
pd.concat([df1,df2], axis=1)
```

## Merge

```python
izq = pd.DataFrame({'key':['k0', 'k1', 'k2', 'k3'],
'A':['A4', 'A5', 'A6', 'A7'],
'B':['B0', 'B1', 'B2', 'B3']})
izq
```

```python
der = pd.DataFrame({'key':['k0', 'k1', 'k2', 'k3'],
'C':['C0', 'C1', 'C2', 'C3'],
'D':['D4', 'D5', 'D6', 'D7']})
der
```

```python
# Derecha: el que está a la derecha del merge (inicial) (al que se la prioridad para mantener sus datos)
# Izquierda: el que está dentro de los paréntesis (el que solo se trae los datos comunes)
# Podemos especificar sobre cual columna se hará el merge

izq.merge(der, on='key')
```

```python
izq = pd.DataFrame({'key':['k0', 'k1', 'k2', 'k3'],
'A':['A4', 'A5', 'A6', 'A7'],
'B':['B0', 'B1', 'B2', 'B3']})
izq

der = pd.DataFrame({'key_2':['k0', 'k1', 'k2', 'k3'],
'C':['C0', 'C1', 'C2', 'C3'],
'D':['D4', 'D5', 'D6', 'D7']})
der
```

```python
# Da error porque no hay una columna llamada 'key' en común

izq.merge(der, on='key')
```

```python
# Una forma de solucionarlo sería así

izq.merge(der, left_on='key', right_on='key_2')
```

```python
izq = pd.DataFrame({'key':['k0', 'k1', 'k2', 'k3'],
'A':['A4', 'A5', 'A6', 'A7'],
'B':['B0', 'B1', 'B2', 'B3']})
izq

der = pd.DataFrame({'key_2':['k0', 'k1', 'k2', np.nan],
'C':['C0', 'C1', 'C2', 'C3'],
'D':['D4', 'D5', 'D6', 'D7']})
der
```

```python
# Como en este caso hay un valor Nan en la fila 3
# No trae esa fila
# Esto no indica que está realizando un INNER JOIN
# no es obligatorio colocar how='inner', ya que es el valor por defecto

izq.merge(der, left_on='key', right_on='key_2')
```

```python
# Hacer un left join

izq.merge(der, left_on='key', right_on='key_2', how='left')
```

```python
# Hacer un right join

izq.merge(der, left_on='key', right_on='key_2', how='right')
```
