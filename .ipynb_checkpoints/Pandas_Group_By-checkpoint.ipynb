```python
import pandas as pd
```

```python
df_books = pd.read_csv('./data/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv')
df_books.head(10)
```

# Groupby

```python
# Agrupar por una columna y su cuenta

df_books.groupby('Author').count()
```

```python
df_books.groupby('Author').mean()
```

```python
df_books.groupby('Author').min()
```

```python
df_books.groupby('Author').max()
```

```python
# En este caso, Author se vuelve el índice del DataFrame

df_books.groupby('Author').sum()
```

```python
# Agrupamos por author y sumamos los datos del registro de William Davis

df_books.groupby('Author').sum().loc['William Davis']
```

```python
df_books.groupby('Author').sum().reset_index()
```

```python
df_books.groupby('Author').mean()
```

```python
# Agrupar por Author con funciones de agregación que nos muestran el mínimo y el máximo por author

df_books.groupby('Author').agg(['min', 'max'])
```

```python
# Agrupar por author usando funciones de agregacion como diccionarios y suma

df_books.groupby('Author').agg({'Reviews':['min', 'max'], 'User Rating': 'sum'})
```

```python
df_books.groupby(['Author','Year']).count()
```

```python
df_books.groupby('Author').agg({'Year':lambda x: [i-2000 for i in x], 'Reviews':lambda x: sum([i**2 for i in x])})
```
