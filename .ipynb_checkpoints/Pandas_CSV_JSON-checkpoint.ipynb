```python
import pandas as pd
```

# CSV

```python
# Leer un .csv .read_csv()
# argumento sep='', tipo de separación: [comas:',', punto y coma ';', espacio:' ', pipe operator: '|']
# argumento header=[number, None], podemos indicar qué fila será los headers, si dejamos None, deja la primera fila
# argumento names=[lista de headers], puedo cambiar los headers de las columnas

df_books = pd.read_csv('./data/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv', sep=',')
df_books
```

```python
# Ejemplo de dataframe usando los argumentos anteriores

df_books = pd.read_csv('./data/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv', sep=',', header=0, names=['Names of Books', 'Authors of Books', 'User Rating', 'Reviews', 'Price', 'Year', 'Genre'])
df_books
```
