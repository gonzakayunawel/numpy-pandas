## Usando Loc

```python
df_books.loc[:]
```

```python
# A diferencia del slicing, loc usa el término final como inclusivo, es decir, si hacemos la selección hasta la row 4
# Esta si aparece dentro de la selección, esto no ocurre con df_books[:4]

df_books.loc[:4]
```

```python
# Con loc puedo combinar el slicing con la búsqueda de labels

df_books.loc[:4, ['Name', 'Author']]
```

```python
# Otro ejemplo de combinación de slicing más labels específicos

df_books.loc[:, ['Reviews']]
```

```python
# Con loc puedo aplicar operaciones numéricas a mis selecciones

df_books.loc[:, ['Reviews']] * -1
```

```python
# Con loc puedo aplicar operaciones numéricas más complejas a mis selecciones
# Para ello nos podemos apoyar con numpy

import numpy as np

# Logaritmo natural

np.log(df_books.loc[:, ['Reviews']])
```

```python
# Logaritmo base 10

np.log10(df_books.loc[:, ['Reviews']])
```

```python
# Raíz cuadrda

np.sqrt(df_books.loc[:, ['Reviews']])
```

```python
# Raíz cúbica

np.cbrt(df_books.loc[:, ['Reviews']])
```

```python
# Media
# Desviación estándar

print('Media igual a: ',np.mean(df_books.loc[:, ['Reviews']]))
print('Desviación estándar igual a: ',np.std(df_books.loc[:, ['Reviews']]))
```

```python
# Podemos crear filtros usando operadores de comparación

df_books.loc[:,['Author']] == 'JJ Smith'
```

## Iloc (iloc)

```python
# Nos permite filtrar y selecionar por columnas, pero, en vez de usar lables
# Se utilizan los índices tanto de las filas o rows como de las columnas o columns
# El primer argumento son los índices de las filas y el segundo las columnas

# Selecionar todas las filas pero solo desde la columna 0 a la 3, para este caso
# La columna de Name sería la columna 0, y la columna 3 no está incluida en la seleción

df_books.iloc[:,0:3]
```

```python
# Podemos acceder a valores específicos y cambiar su valor al igual que con loc

df_books.iloc[1,3] * -1
```

```python
# Podemos usar el slicing para acceder a distintas visualizaciones
# Acá vemos desde la fila 0 a la 1 (la 1 no se incluye)
# Y la columa 0 a la 1 (la 1 no se incluye)

df_books.iloc[:1,:1]
```

```python
# Podemos usar el slicing para acceder a distintas visualizaciones
# Acá vemos desde la fila 0 a la 1
# Y desde la columna 0
df_books.iloc[:1,0:7]
```
